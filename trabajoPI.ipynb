```python
def extract_id(value): #desasnidar
    if pd.isnull(value):
        return np.nan
    try:
        collection_dict = ast.literal_eval(value)
        if isinstance(collection_dict, dict) and 'id' in collection_dict:
            return collection_dict['id']
        else:
            return np.nan
    except (ValueError, SyntaxError):
        return np.nan
```

```python
def limpiar_extraer(languages):
    languages = ast.literal_eval(languages) #convierto a lista de dicc
    data_limpio = []
    for lang in languages:
        if isinstance(lang, dict) and 'iso_639_1' in lang and 'name' in lang:#verifico el dicc y las claves
            data_limpio.append((lang['iso_639_1'], lang['name']))
    return data_limpio
belongs_to_collection_list = []
id_list = []
language_abrev = []
language_movie = []

for idx, row in df_languages_copy.iterrows():
    clean_languages = limpiar_extraer(row['spoken_languages'])#usamos la func
    for lang in clean_languages:
        belongs_to_collection_list.append(row['belongs_to_collection'])#duplicamos por la cant de leng
        id_list.append(row['id'])
        language_abrev.append(lang[0])
        language_movie.append(lang[1])

df_limpio = pd.DataFrame({ #nuevo df
    'belongs_to_collection': belongs_to_collection_list,
    'id': id_list,
    'language_abrev': language_abrev,
    'language_movie': language_movie })

display(df_limpio)
```
